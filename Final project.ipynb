# METHODS
The methodology employed in this study comprises a multi-faceted approach that integrates data analysis and visualization techniques to comprehensively explore the optimization of flight routes. Initially, the dataset undergoes rigorous cleaning processes to ensure accuracy and reliability, including column splitting and the removal of redundant information. Geographical coordinates for origin and destination cities are then obtained using the Nominatim geocoding service, providing the spatial context necessary for route visualization.

Subsequently, various libraries such as Matplotlib, Cartopy, Folium, and Bokeh are utilized to visualize the data in diverse formats. Matplotlib and Cartopy facilitate the plotting of flight routes on static maps, offering a geographical perspective of route distributions within the United States. Folium, on the other hand, is employed to create interactive maps with markers representing origin cities and the total number of passengers, enhancing the user's engagement and understanding of the spatial distribution of airports and route popularity. Additionally, Bokeh enables the visualization of airport locations on a Mercator projection map, providing insights into the spatial distribution of airports and their relationships with flight routes.

Furthermore, route optimization is facilitated by calculating distances between origin and destination airports using the Haversine formula—a technique employed to estimate distances between two points on a sphere. These calculated distances enable the identification of key routes and hubs, informing strategic decisions regarding route planning and optimization in the airline industry.

Overall, this methodological framework encompasses a blend of data preprocessing, geospatial analysis, and visualization techniques to unravel insights that inform strategic decision-making processes aimed at enhancing operational efficiency, cost-effectiveness, and passenger satisfaction within the aviation sector.

# Loading and cleaning the data


```python
file_path = '/Users/beryloch/Desktop/Final project /data visual data/airline dataset/Top 1000 2015 US Airline Route Segments by Volume.xlsx'
df = pd.read_excel(file_path)
df[['CARRIER_NAME', 'ORIGIN_CITY_NAME', 'DEST_CITY_NAME']]  = df['CARRIER_NAME-ORIGIN_CITY_NAME-DEST_CITY_NAME'].str.split('-', expand=True)

def remove_total_total(city_name):
    if city_name and isinstance(city_name, str):  
        if "Total Total" in city_name:
            city_name = city_name.replace("Total Total", "").strip()
    return city_name


df['DEST_CITY_NAME'] = df['DEST_CITY_NAME'].apply(remove_total_total)
df[['CARRIER_NAME', 'ORIGIN_CITY_NAME', 'DEST_CITY_NAME']]

combined_df = pd.concat([df,df[['CARRIER_NAME', 'ORIGIN_CITY_NAME', 'DEST_CITY_NAME']] ], axis=1)
new_df = combined_df[['PASSENGERS', 'Passenger Miles', 'CARRIER_NAME', 'ORIGIN_CITY_NAME', 'DEST_CITY_NAME']]
new_df



geolocator = Nominatim(user_agent="city_locator")
city_coordinates = {}
def get_coordinates(city_name):
    if city_name not in city_coordinates:
        location = geolocator.geocode(city_name)
        if location:
            city_coordinates[city_name] = (location.latitude, location.longitude)
        else:
            city_coordinates[city_name] = (None, None)
    return city_coordinates[city_name]


df['ORIGIN_LATITUDE'] = df['ORIGIN_CITY_NAME'].apply(lambda x: get_coordinates(x)[0])
df['ORIGIN_LONGITUDE'] = df['ORIGIN_CITY_NAME'].apply(lambda x: get_coordinates(x)[1])
df['DEST_LATITUDE'] = df['DEST_CITY_NAME'].apply(lambda x: get_coordinates(x)[0])
df['DEST_LONGITUDE'] = df['DEST_CITY_NAME'].apply(lambda x: get_coordinates(x)[1])

df.to_csv("updated_data.csv", index=False)
file_path = '/Users/beryloch/Desktop/Final project /data visual data/updated_data.csv'
newdata = pd.read_csv(file_path)
newdata.head(100)
```

# Geospatial Analysis 
 
 Visualizing the data on map to understand the geographical distribution of passengers traffic and identify popular destinations and understands the regions.

```python
# Group the data by origin city and calculate the total number of passengers for each city
origin_passengers = newdata.groupby('ORIGIN_CITY_NAME').agg({'PASSENGERS': 'sum', 'ORIGIN_LATITUDE': 'mean', 'ORIGIN_LONGITUDE': 'mean'}).reset_index()

# Create a map centered on the United States (you can adjust the center and zoom level based on your data)
m = folium.Map(location=[37.0902, -95.7129], zoom_start=7)

# Add markers for each origin city with the total number of passengers as the popup
for index, row in origin_passengers.iterrows():
    folium.CircleMarker(
        location=[row['ORIGIN_LATITUDE'], row['ORIGIN_LONGITUDE']],
        radius=row['PASSENGERS'] / 100000,  # Adjust the radius for better visualization
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"{row['ORIGIN_CITY_NAME']}<br>Total Passengers: {row['PASSENGERS']}"
    ).add_to(m)
m
```

```python
df1 = pd.read_csv("/Users/beryloch/Desktop/Final project /data visual data/updated_data.csv")  

def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = 6371 * c  
    return distance

distances = []
for index, row in df.iterrows():
    origin_lat = row['ORIGIN_LATITUDE']
    origin_lon = row['ORIGIN_LONGITUDE']
    dest_lat = row['DEST_LATITUDE']
    dest_lon = row['DEST_LONGITUDE']
    
    distance = haversine(origin_lat, origin_lon, dest_lat, dest_lon)
    distances.append(distance)

newdata['DISTANCE_KM'] = distances
newdata
```

```python

plt.figure(figsize=(10, 8))
ax = plt.axes(projection=ccrs.AlbersEqualArea(central_longitude=-96, central_latitude=39))
ax.set_extent([-125, -65, 20, 50], ccrs.Geodetic())


ax.add_feature(cfeature.STATES.with_scale('10m'), edgecolor='black', linestyle='-', linewidth=0.5)
ax.add_feature(cfeature.LAND.with_scale('10m'), facecolor='lightgray')
ax.add_feature(cfeature.BORDERS.with_scale('10m'), linestyle='-', linewidth=0.5)

for index, row in newdata.iterrows():
    origin_lat = row['ORIGIN_LATITUDE']
    origin_lon = row['ORIGIN_LONGITUDE']
    dest_lat = row['DEST_LATITUDE']
    dest_lon = row['DEST_LONGITUDE']
    ax.plot([origin_lon, dest_lon], [origin_lat, dest_lat], transform=ccrs.PlateCarree(), color='blue', linewidth = 0.1)

plt.title('Flight Routes within US')
plt.show()
```

```python
route_passengers = newdata.groupby(['ORIGIN_CITY_NAME', 'DEST_CITY_NAME']).agg({'PASSENGERS': 'sum'}).reset_index()
route_passenger_miles = newdata.groupby(['ORIGIN_CITY_NAME', 'DEST_CITY_NAME']).agg({'Passenger Miles': 'sum'}).reset_index()

most_popular_routes_passengers = route_passengers.sort_values(by='PASSENGERS', ascending=False).head(10)
most_popular_routes_passenger_miles = route_passenger_miles.sort_values(by='Passenger Miles', ascending=False).head(10)


plt.figure(figsize=(10, 6))
plt.barh(most_popular_routes_passengers['ORIGIN_CITY_NAME'] + ' to ' + most_popular_routes_passengers['DEST_CITY_NAME'], most_popular_routes_passengers['PASSENGERS'])
plt.xlabel('Number of Passengers')
plt.ylabel('Route')
plt.title('Top  Most Popular Routes by Number of Passengers')
plt.gca().invert_yaxis()  
plt.show()

plt.figure(figsize=(10, 6))
plt.barh(most_popular_routes_passenger_miles['ORIGIN_CITY_NAME'] + ' to ' + most_popular_routes_passenger_miles['DEST_CITY_NAME'], most_popular_routes_passenger_miles['Passenger Miles'])
plt.xlabel('Passenger Miles')
plt.ylabel('Route')
plt.title('Top Most Popular Routes by Passenger Miles')
plt.gca().invert_yaxis()  
plt.show()
```
